# Probabilistic Robotics Defined
Most of the time, robotics is considered to be a *cyber-physical system*. That is, a system in which computer programs drive physical manipulators. Our concern, however, is how these control programs receive sensory data and subsequently act upon it.

One such school of modeling derives the following: Each component in the system can be modeled mathematically with "reasonable" accuracy. Thus, the control of a robot relies upon the integration of these models in the form of feed-forward or feed-back control. This paradigm has some design issues that it must overcome.

First, we must consider that models in and of themselves are inaccurate: every model omits certain details (biases), and is limited by being a deterministic approximation of something that is inherently nondeterministic. The second is that as systems become increasingly complex, managing these inaccuracies becomes impossible. In engineering physical devices, there are *tollerance* specifications, all with the idea (hope really) that the cumulative errors ballance out. But, as a system moves, and (worse yet) lacks a global reference point at every instant, the sum of errors becomes divergent. Sources of these errors are not as simple as *sensor noise*, but also include manufactoring defects, inaccuracies with ground-truth, clock drift, and conflicting sensory information.

*Probabilistic Robotics* aims to address that. Rather than being limited to models which need refreshed in the face of cumulative error, we can represent ambiguity and belief as part of the model. Interestingly enough, this makes for far more robust systems, as the probabilistic combinations of these error models simply becomes another constraint to optimize.

There are consequences to this modeling paradigm. First is the computational complexity: in comparison to models-based robotics (which only computes a single solution), this paradigm computes the entire space of solutions, and is thus constrained by our discretization design. The second is that we sacrifice efficient solutions in favor of solutions with the highest overall probability of success. Consider a robot crossing a large room reaching a destination point -- following a wall with a handful of landmarks allows for accurate course-correction instead of driving through free-space without aligning guides.
